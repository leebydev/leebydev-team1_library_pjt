```python
import pandas as pd
import numpy as np
import random
```

```python
user_info = pd.read_csv('../data2/user_info_80000.csv')
book_info = pd.read_csv('../data2/book_info_80000.csv')
```

```python
user_info = user_info.reset_index(drop=True)
display(user_info.head())
book_info = book_info.reset_index(drop=True)
display(book_info.head())
```

```python
display(user_info.head())
```

```python
user_info2 = user_info.iloc[:, 1:]
book_info2 = book_info.iloc[:, 1:]
#display(user_info2.head())
#display(book_info2.head())
```

```python
df = pd.concat([user_info2, book_info2], axis=1)
display(df.head())
len(df)
```

```python
df2 = df.assign(sex=np.nan)
display(df2)
```

```python
df2.iloc[:40000, -1] = 0
df2.iloc[40000:80000, -1] = 1
```
