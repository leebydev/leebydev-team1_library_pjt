```python
import pandas as pd
```

```python
df = pd.read_csv('../data2/library_learning_dataset.csv')
df.info()
display(df.head())
```

```python
df[['kdc_class_no','age','sex']] = df[['kdc_class_no','age','sex']].astype(int)
```

```python
display(df.head())
```

```python
loan = df['user_key'].value_counts()
```

```python
loan
```

```python
loan = loan.to_frame(name='T')
loan.reset_index(drop=False, inplace=True)
display(loan)
```

```python
loan.rename(columns = {'index':'user_key'}, inplace=True)
display(loan)
```

```python
loan_5 = loan[loan['T'] > 5]
```

```python
df_5 = pd.merge(df, loan_5, on=['user_key'], how='inner')
display(df_5)
```

```python
df_5.to_csv('../data2/library_learning_dataset_loan5.csv', index = False)
```
